# Data cleaning

```python
punesoil_data.isnull().sum()   #checking missing values
```

```python
df1=punesoil_data.drop(['Unnamed: 1','Unnamed: 11'],axis='columns')  # Deliting Two missing columns
print(df1)
```

```python
df1.info()      # Information of given data
```

```python
df1['Mn']=df1['Mn'].fillna(df1['Mn'].mean())
df1['N']=df1['N'].fillna(df1['N'].mean())
df1['P']=df1['P'].fillna(df1['P'].mean())
df1['K']=df1['K'].fillna(df1['K'].mean())
df1['Zn']=df1['Zn'].fillna(df1['Zn'].mean())
df1['B']=df1['B'].fillna(df1['B'].mean())
df1['Cu']=df1['Cu'].fillna(df1['Cu'].mean())
df1['Fe']=df1['Fe'].fillna(df1['Fe'].mean())
df1['S']=df1['S'].fillna(df1['S'].mean())
df1['pH']=df1['pH'].fillna(df1['pH'].mean())
df1['EC']=df1['EC'].fillna(df1['EC'].mean())
df1['OC']=df1['OC'].fillna(df1['OC'].mean())   # Replacing missing values by its mean
```
